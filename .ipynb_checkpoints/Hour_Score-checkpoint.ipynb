```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
```

```python
#saate motaleE
X_train = np.random.uniform(0,12,300)
#nomre daneshjuyan
Y_train = X_train + np.random.uniform(0,8,300)

plt.scatter(X_train, Y_train)
plt.xlabel('Hour')
plt.ylabel('Score')
```

```python
X = X_train.reshape((300,1))
Y = Y_train.reshape((300,1))
Y.shape
```

```python
# correct m
m = np.matmul(inv(np.matmul(X.T, X)) , np.matmul(X.T, Y))
m = m[0,0]
m
```

```python
Y_pred  = X * m
```

```python
# False_m
m1 = 0.8
y_pred1 = X * m1

m2 = 1.9
y_pred2 = X * m2

m3 = 2.8
y_pred3 = X * m3
```

```python
#test 1
plt.plot(X , y_pred1, c='yellow')
plt.scatter(X, Y)
```

```python
#test 2
plt.plot(X , y_pred2, c='green')
plt.scatter(X, Y)
```

```python
#test 3
plt.plot(X , y_pred3, c='pink')
plt.scatter(X, Y)
```

```python
plt.scatter(X, Y, c='blue', label='data')
plt.plot(X , y_pred1, c='yellow', label='test 1')
plt.plot(X , y_pred2, c='green', label='test 2')
plt.plot(X , y_pred3, c='pink', label='test 3')
plt.plot(X ,Y_pred, c='red', label='linear least squares line')

plt.legend()
plt.show()
```
